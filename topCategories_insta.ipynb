# filtering out audience in united nation

```python
us_df = df[df["Audience Country"] == "United States"].copy()
us_df
```

```python
us_df['Category'] = us_df['Category'].str.split('|')
us_df_exploded = us_df.explode('Category')
```

```python
us_df_exploded
```

```python
us_df_exploded.head()
```

```python
top_categories = us_df_exploded['Category'].value_counts().head(10)
top_categories
```

# visualization of this data we got

```python
#creating histogram for this data

plt.figure(figsize = (12,3))
sns.histplot( us_df_exploded["Followers"], bins = 30, kde = True, color = "skyblue")
plt.title('Distribution of Followers for U.S. Instagram Influencers', fontsize = 16)
plt.xlabel("no. of followers", fontsize = 14)
plt.ylabel("frequency", fontsize = 14)
plt.show()
```

```python
#creating a piechart
plt.figure(figsize=(14, 8))
plt.pie(top_categories.values, labels=top_categories.index, autopct='%1.1f%%', startangle=90 , shadow = True, colors=sns.color_palette('pastel'))
plt.title('Top 10 U.S. Instagram Influencer Categories', fontsize=16)
plt.show()
```
